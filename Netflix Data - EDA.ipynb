### Find Missing Values

```python
# number of missing values
df.isnull().sum()
```

```python
# percentage of missing values
null_perc = round(df.isnull().sum() / df.shape[0] * 100, 2).to_dict()
for attribute in null_perc:
    if null_perc[attribute] > 0:
        print(f"{attribute}: {null_perc[attribute]}%")
```

### Number of TV Shows vs Movies

```python
df.type.unique()
```

```python
plt.pie(df.type.value_counts(), labels = ["Movies", "TV Show'"], autopct = "%1.2f%%", colors = ["#D00", "#AA0"])
plt.title("TV Shows vs Movies: Count")
plt.show()
```

From the above pie chart we can observe that about *70%* of the content on *Netflix* are *Movies*
